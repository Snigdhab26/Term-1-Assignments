```python
earthquake.describe()
```

Analysis:
1. Latitude : From latitude and longitude we can understand the geographic position of the earthquake. So,if the latitude is negative then earthquake is in the Southern hemisphere and otherwise in the northern hemisphere.From the  observations of this variable, it can be observed that atleast 50% of the earthquakes happened in the southern hemisphere ( as latitude is negative).
2. Longitude : From latitude and longitude we can understand the geographic position of the earthquake. So,if the longitude is negative then western hemisphere & otherwise it is in eastern hemisphere .From the observations of this variable, it can be observed that atleast 25% of the earthquakes happened in the western hemisphere ( as longitude is negative ) but most of the  earthquakes happened in the eastern hemisphere ( as longitude is positive)

So, it can be deduce from the data that most prone area is South Eastern region of the Hemisphere.
2. Depth: This depicts the depth of earthquake. It is a numerical data that can provide us with distribution of earthquake depth occuring over the years since 1965. The distribution is rightly skewed as Mean > Median. 
3. Depth Error: There are missing values in this variable as count < 23412.As it can be statistically important variable imputing the variable should be a good option. Imputing the variable with zero as drastically pull down mean and SD . As data missing is way more than 30% imputing it will not provide desired anaylysis . Hence droping the column .
4. Depth Sesimic Stations : This is a numerical data as it represents number of stations.It can be seen many data is missing. So imputing can result into incorrect analysis. Hence this column too.
5. Magnitued : This variable seems to be normally distributed with not much variance in the data than.
6. Magnitude Error , Magnitude Seismic stations ,Azimuthal Gap , Horizontal Distance and Horizontal Error: These variables has large chunk of data missing . Analysing them by imputing will not be representative of the actual distribution. Hence droping these columns .
7. Root Mean Square:This is a numerical data and has some missing values. Also, the Mean is similar to median. Hence, imputing this variable median will be safer option.

```python
new_earthquake = earthquake.drop(columns = ['Depth Error','Depth Seismic Stations','Magnitude Error','Magnitude Seismic Stations','Azimuthal Gap','Horizontal Distance','Horizontal Error'])
new_earthquake

new_earthquake['Root Mean Square'].fillna(new_earthquake['Root Mean Square'].median(),inplace=True)
new_earthquake.describe()
```

```python
new_earthquake.info()
```

There is no more missing values except for Magnitude Type and its a categorical value  will not be imputing it.

```python
new_earthquake.describe()
```

Checking the correlation between below variables:
1. Depth v/s Magnitude
2. Depth v/s Root Mean Square
3. Year v/s Magnitude

```python
#Depth v/s Magnitude
figure = plt.figure(figsize =(15,8))
x=new_earthquake['Depth']
y=new_earthquake['Magnitude']
plt.scatter(x,y)
plt.xlabel("Depth")
plt.ylabel("Magnitude")
plt.title("Depth v/s Magnitude")
plt.show()
```

From the graph, it can be observed that earthquakes around the depth 300-400km has lower magnitude whereas the magnitude is higher at the lower and higher depths.

```python
#Depth v/s Root Mean Square
figure = plt.figure(figsize =(15,8))
x=new_earthquake['Depth']
y=new_earthquake['Root Mean Square']
plt.scatter(x,y)
plt.xlabel("Depth")
plt.ylabel("Root Mean Square")
plt.title("Depth v/s Root Mean Square")
plt.show()
```

From the graph, the Root Mean Square amplitude is approx constant across the depth with few outliers on the lower depth end.

```python
#Year v/s Magnitude
from datetime import datetime 
new_earthquake['Date'] = pd.to_datetime(new_earthquake['Date'],utc=True)
new_earthquake['Earthquake_Year']=pd.DatetimeIndex(new_earthquake['Date']).year

figure = plt.figure(figsize=(15,8))
x=new_earthquake['Earthquake_Year']
y=new_earthquake['Magnitude']

plt.scatter(x,y)
plt.xlabel("Year")
plt.ylabel("Magnitude")
plt.title("Year v/s Magnitude")
plt.show()
```

Majority of earthquakes are in the magnitude range 5-7 .Very few lies above 8.

To conduct data analysis on the Indian Cities dataset, we first need to import the data and determine what questions we would like to address. Here are some potential questions that could be answered through this analysis:

1. What is the distribution of the population across different states in India?
2. What is the average sex ratio across all cities in India?
3. Which cities have the highest and lowest literacy rate in each state?
3. What is the distribution of the literacy rate across different states in India?
4. What is the distribution of the number of graduates across?

```python
indian_city = pd.read_csv('Indian_cities.csv')
display(indian_city)
indian_city.describe()
```

```python
indian_city.info()
```

There is no null record present

```python
# Distribution of the population across different states in India
import matplotlib.pyplot as plt

population_dist = indian_city.groupby('state_name')['population_male','population_female'].sum()

population_dist.plot(kind='bar',stacked=False, figsize=(15,8))

plt.title("Distribution of Population Across Different States in India")
plt.show()

```

Above graph shows that highest population is in Mahrahstra. Also,in all the states the male ratio is higher than the female ratio.

```python
#Average sex ratio across all cities in India


sex_ratio_dist = indian_city['sex_ratio'].mean()
sex_ratio_dist
```

```python
#Which cities have the highest and lowest literacy rate in each state?


# group by state name and find the city with the highest effective literacy rate total in each state
state_literacy = indian_city.groupby('state_name')
city_with_highest_literacy = state_literacy.apply(lambda x: x[x.effective_literacy_rate_total == x.effective_literacy_rate_total.max()])

# display the city with the highest effective literacy rate total in each state
print(city_with_highest_literacy[['state_name', 'name_of_city', 'effective_literacy_rate_total']])

```

```python
#distribution of the literacy rate across different states in India
states_India = indian_city['state_name']
states_India

literacy_dist = indian_city.groupby('state_name')['literates_male','literates_female'].sum()


literacy_dist.plot(kind='bar',stacked=False, figsize=(15,8))

plt.title("Distribution of the literacy rate across different states in India")
plt.show()
```

From the graph, it can be deduced that literacy rate in maharashtra is highest.Also, in all the states the male literacy rate is higher than the female literacy rate.

```python
#Distribution of the number of graduates across
states_India = indian_city['state_name']
states_India

graduate_dist = indian_city.groupby('state_name')['male_graduates','female_graduates'].sum()


graduate_dist.plot(kind='bar',stacked=False, figsize=(15,8))

plt.title("Distribution of the number of graduates across")
plt.show()
```

From the graph, it can be deduced that number of graduates in maharashtra is highest, while that in manipur , meghalaya ,mizoram and Nagaland is comparatively very less.As from the previous graph we observed that literacy is highest in Maharashtra and hence will be checking the correlation between number of graduates and literacy rate.
