## 1.Product name


```python

# Extracting product name from below mentioned classname
prod_name = soup.select(".srp-results .s-item__title")
product = [pt.get_text().strip() for pt in prod_name]


product
#product.index('Nautical round table rope designer natural wooden top coffee tea side table item')
```

## 2.Price


```python
# Extracting product price from the page and assigning to a list

prod_price = soup.select(".srp-results .s-item__price")
prod_price_range =[pt.get_text().strip() for pt in prod_price]
prod_price_range

#prod_price_range.index('$299.00')
```

## 3.Shipping cost


```python
# Extracting shipping cost from the page and assigning to a list

shipping_cost = soup.select(".srp-results .s-item__shipping")

product_shipping_cost = [pt.get_text().strip() for pt in shipping_cost]
product_shipping_cost
```

## 4.Image URL


```python
## Extracting product image from the page and assigning to a list


block = soup.find('ul', attrs = {'class':'srp-results srp-list clearfix'})
product_items = block.find_all('div',attrs = {'class':'s-item__wrapper clearfix'})
image = block.find_all('div', class_='s-item__wrapper clearfix')
prod_images =[]
for i in image:
    prod_image_link = i.find('img')
    prod_images.append(prod_image_link.get('src'))
   
```

```python
prod_images
#prod_images.index('https://i.ebayimg.com/thumbs/images/g/gfcAAOSwGS1iLEWp/s-l225.jpg')
```

## 5.Number of watchers


```python
# Extracting number of watchers from the page and assigning to a list

No_Watcher =[]
for i in soup.find_all('div', class_='s-item__info clearfix'):
    Watchers = i.find('span',class_='s-item__dynamic s-item__watchCountTotal')
    if Watchers:
        #print(Watchers.get_text())
         No_Watcher.append(Watchers.get_text())
    else:
        No_Watcher.append('No Watchers')
        
No_Watcher
```

## 6.Product condition (Brand new, Opened Box or Pre-owned)

```python
prod_cond = soup.select(".srp-results .SECONDARY_INFO")

product_condition = [pt.get_text().strip() for pt in prod_cond]
#del product[0]
product_condition
```

```python
import csv

rows = zip(product, prod_price_range,product_shipping_cost, prod_images,No_Watcher, product_condition)
head = ['Product name', 'Price', 'Shipping cost','Image URL' ,'Number of Watchers', 'Product condition']
  
with open('tablelist.tsv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f, delimiter='\t') 
    writer.writerow(head)
    for row in rows:
        writer.writerow(row)
        


#zipping all the list int one and then writing data into a tsv file       
        
```

```python
Header = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
       'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
       'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
       'Accept-Encoding': 'none',
       'Accept-Language': 'en-US,en;q=0.8',
       'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
       'Connection': 'keep-alive'}

page1 = requests.get("https://www.ebay.com/sch/i.html?_nkw=table", headers=Header)
page1
```

```python
page1.content
from bs4 import BeautifulSoup
soup = BeautifulSoup(page1.content, 'html.parser')
```

```python
#Listing Product URLs so as to loop each product to find other details
Prod_Box = soup.find('ul',class_='srp-results srp-list clearfix' )
products_class = Prod_Box.find_all('div',class_='s-item__wrapper clearfix')

prod_url=prod_url = [product.find('a')['href'] for product in products_class]
prod_url 
```

```python
#Defining function that will provide details for each product . This function will be used later in a loop to execute for each product link

def ebay_product(link):
    
    global df   
        
    #link ='https://www.ebay.com/itm/125377736077?hash=item1d31186d8d:g:GekAAOSwS2lir~AU&amdata=enc%3AAQAHAAAA0EV9z92PhqZbr4P948VF3jwuWWJbqqs7STpTCLBPU4Mv%2BdO6Z3ibexCyUUkw84MHcH8jUw2PDRo4tMt9XUFuObR6GvQ%2BsEy6lWp%2BkV2lWtGQqX4eZWNbYI5BwudyBXknk9gJVVKUo0B38bS710b9wF1SwUlSAJoOFO4HVyfYOGZEhl4EHrqAJqzZ3x6qZ8e7TtFryii1fM7Bnp5So5qK5BjmT9D1fyFL7vXUjPSJHvP5n0%2BFe%2FP8a%2Fpow4CnNxNl%2B9zlFhqr9AEJaJ%2FWBLmVNGU%3D%7Ctkp%3ABFBMsu6e9uZh'
    
    page2 = requests.get(link,headers=Header)
        
    soup2 = BeautifulSoup(page2.text, 'html')
    
    #Creating an empty dict to columns that later will be combined in dataframe
    dict1 = {} # This dictionary will hold data like product name , payment modes available for each product and shipping details and price
    dict2 = {} # This dictionary will hold data 
    
    LeftPanel = soup2.find('div',id='LeftSummaryPanel') # locating the element LeftSummaryPanel to extract information
    
    prodname = LeftPanel.find('span',class_='ux-textspans ux-textspans--BOLD').get_text()
    dict1['Product Name'] = prodname # extracting product name from the passed product link
    
    
    return_policy = LeftPanel.select_one('div.vim.x-returns-minview div.ux-labels-values__values.col-9').get_text()
    dict1['Returns Policy'] = return_policy # extracting return policy of the passed product link
    
    
    bid_price = LeftPanel.find_all('div',class_='x-buybox__section')

    for price in bid_price:
        if price.find('div',class_='x-label'):
            label = price.find('div',class_='x-label').get_text().replace(":","")

        if price.find('div',class_='x-price-primary'):
            value = price.find('div',class_='x-price-primary').get_text()
            dict1[label] = value
 
    # extracting starting bid of the product
    
    
    srp_rows = LeftPanel.find_all('div', class_='ux-layout-section__row')
    for row in srp_rows:
        label = row.find('div', class_='ux-labels-values__labels-content').get_text()
        content = row.find('div', class_='ux-labels-values__values-content').get_text()
    # extracting bid prices and assigning to dictionary
        if label == 'Shipping:':
            substring = 'See detailsfor shipping'
            dict1['Shipping cost'] = content[0:content.find(substring)].replace("\xa0", ".") if substring in content else content
        elif label == 'Delivery:':
            substring = 'Estimated delivery dates - opens in a new window'
            dict1['Estimated delivery date'] = content[0:content.find(substring)] if substring in content else content
    #checking if the labal has value SHipping or delivery to extract Shipping cost and estimated delivery date
    
    payment_modes = LeftPanel.find_all('span', role="img")
    modeslist = [p.get("title") for p in payment_modes if p.get("title") is not None]
    dict1['Payment Modes'] = ','.join(modeslist)
    
################################################################################################

    tab = soup2.find_all('div', class_='vi-VR-tabCnt')
    rows = tab[0].find_all('div', class_='ux-layout-section__row')

    head = []
    val = []

    # Fetching eBay number
    head.append(rows[1].get_text().split(":")[0])
    val.append(rows[1].get_text().split(":")[1])

    for row in rows:
        lab = row.find_all('div', class_='ux-labels-values__labels')
        cont = row.find_all('div', class_='ux-labels-values__values')

        for l, c in zip(lab, cont):
            head.append(l.get_text().replace(":", ""))
            field_cont = c.get_text().replace("opens in a new window or tab", "").strip()
            val.append(field_cont)

        for i in range(len(head)):
            dict2[head[i]] = val[i]

    # Clear off the lists
    head.clear()
    val.clear()

   # Fetching "Shipping, returns and payments" section
    srp_section = tab[1].find('div', {'id': 'vi-ship-maincntr'})

   # "Item Location" and "Ships To" information
    info = srp_section.find_all('div', class_='ux-layout-section__row')

    for i in info:
        dict_key = i.find('div', class_='ux-labels-values__labels col-6')
        dict_val = i.find('div', class_='ux-labels-values__values col-6')

        if dict_key and dict_val:
            dict_key = dict_key.get_text().replace("\xa0", "").replace(":", "")
            dict_val = dict_val.get_text().strip()
            dict2[dict_key] = dict_val

    final = {**dict1,**dict2}
    columns = ['Brand','Condition','eBay item number','Material','Type','Item location','Ships to','Product Name','Price','Starting bid','Payment Modes','Returns Policy','Estimated delivery date','Shipping cost','Color']

    #Handling Missing values
    for col in columns:
        if col not in final.keys():
            final[col] = 'Not Found'

    #Selecting required key-value pairs in the dictionary
    final = {key:final[key] for key in columns}   
    
    df = df.append(final,ignore_index=True)
```

```python
df = pd.DataFrame()

for url in prod_url:
    ebay_product(url)
    
df = df[['Product Name','Price','Starting bid','Payment Modes','Returns Policy','Brand','Material','Color','Type','Condition','eBay item number','Ships to','Item location','Shipping cost','Estimated delivery date']]    
df.to_csv("tableDetails.tsv",sep='\t')

df

```
